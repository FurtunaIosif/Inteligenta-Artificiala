```python
#Exemplu de invatare hebbiana

import numpy as np

w = [1,-1,0,0.5]

net = 0
x = 0        #Numarul de iteratii
c = 1        #Rata de invatare

while x < 10:
    
    x1 = [1,-2,1.5,0]
    x2 = [1,-0.5,-2,-1.5]
    x3 = [0,1,-1,1.5]
    

    print("Se aplica x1:")
    
    net = np.dot(w,x1)
    fnet = [y * np.sign(net) for y in x1]
    w = np.add(w,fnet)
    print(w,"\n")


    print("Se aplica x2:")
    
    net = np.dot(w,x2)
    fnet = [y * np.sign(net) for y in x2]
    w = np.add(w,fnet)
    print(w,"\n")


    print("Se aplica x3:")
    
    net = np.dot(w,x3)
    fnet = [y * np.sign(net) for y in x3]
    w = np.add(w,fnet)
    print(w,"\n")
    
    x = x + 1
```
