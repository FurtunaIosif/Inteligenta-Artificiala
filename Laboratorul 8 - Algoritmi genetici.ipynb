```python
import numpy as np
import random

S = 36 #suma la care vrem sa ajungem
P = 360 #produsul la care vrem sa aungem
Lista = [] 

m = np.random.randint(2, size = (50,10)) #matricea cu elemente 0 si 1 random

a = 0
while a < 20:
    for i in range (0,50):
        x = 1 #counter
        S1 = 0 #suma initiala
        P1 = 1 #produsul initial
        
        for j in range (0,10):
            if m[i][j] == 0 :
                S1 += x
                x = x + 1
            else:
                P1 *= x
                x = x + 1
        F = (abs(S1 - S) / S) + (abs(P1 - P) / P) #Fitness function
        Lista.append(F) # adaug in lista functia fitness
    print("\n\nLista inainte de ranking ------>\n\n",Lista)
    
    for num in range(len(Lista)-1,0,-1): #fac ranking prin metoda bulelor
        for i in range(num):
            if Lista[i]>Lista[i+1]:  # ranking in lista cu functii fitness
                temp = Lista[i]
                Lista[i] = Lista[i+1]
                Lista[i+1] = temp
                for j in range(0,10): # ranking linii matrice
                    temp1 = m[i][j]
                    temp2= m[i+1][j]
                    m[i][j] = temp2
                    m[i+1][j] = temp1
    print("\n\nLista dupa ranking ------->\n\n",Lista)
    del Lista[:]
    k = 46
    for i in range(0,4):  
        for j in range(0,10):
            m[k][j] = m[i][j]
        k += 1
    n = random.sample(range(2 , 49), 20)
    i = 0
    while i < 19: #aici fac crossover
        for j in range (5,10):
            temp1 = m[n[i]][j]
            temp2 = m[n[i+1]][j]
            m[n[i]][j] = temp2
            m[n[i+1]][j] = temp1
        i = i + 2

    r1 = random.sample(range(2 , 49), 5)#5 numere random care vor semnifica liniile pe care se va face mutatia
    r2 = random.sample(range(0 , 8), 5) #5 numere random care vor semnifica coloanele pe care se va face mutati

    i = 0
    j = 0
    while i < 5:
        temp1 = m[r1[i]][r2[j]]
        temp2 = m[r1[i] + 1][[r2[j]]]
        m[r1[i]][r2[j]] = temp2
        m[r1[i] + 1][[r2[j]]] = temp1
        i = i + 1
        j = j + 1
    a += 1
print(m)
```
