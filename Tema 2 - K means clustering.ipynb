```python
import numpy as np
import random
import math

x = [45,50,40,55,200,48,195,43,190] #coordonata "x" a punctelor
y = [85,43,80,42,43,40,41,87,40] #coordonata "y" a punctelor

centr1 = [0,0]
centr2 = [0,0]
centr3 = [0,0]

k  = 3 #numarul de clustere

cl1x = [] #lista cu coordonatele "x" a punctelor din clusterul 1
cl1y = [] #lista cu coordonatele "y" a punctelor din clusterul 1

cl2x = [] #lista cu coordonatele "x" a punctelor din clusterul 2
cl2y = [] #lista cu coordonatele "y" a punctelor din clusterul 2

cl3x = [] #lista cu coordonatele "x" a punctelor din clusterul 3
cl3y = [] #lista cu coordonatele "y" a punctelor din clusterul 3

Med1 = [] #media punctelor din clusterul 1
Med2 = [] #media punctelor din clusterul 2
Med3 = [] #media punctelor din clusterul 3


n = random.sample(range(0 , 8), 3) #pozitiile pe care se afla primii 3 centroizi

print("Vectorul random:",n)

for i in range (3):
    print("Centrul de greutate",i + 1, "este:", "[" ,x[n[i]],y[n[i]],"]")

    
for i in range (0,9):
    d1 = math.sqrt((x[i] - x[n[0]])**2 +(y[i] - y[n[0]])**2) #distanta de la punctul de pe pozitia "i" la primul centroid
    d2 = math.sqrt((x[i] - x[n[1]])**2 +(y[i] - y[n[1]])**2) #distanta de la punctul de pe pozitia "i" la al doilea centroid
    d3 = math.sqrt((x[i] - x[n[2]])**2 +(y[i] - y[n[2]])**2) #distanta de la punctul de pe pozitia "i" la al treilea centroid
    if d1 < d2 and d1 < d3:
        print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 1")
        cl1x.append(x[i])
        cl1y.append(y[i])
    elif d2 < d1 and d2 < d3:
        print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 2")
        cl2x.append(x[i])
        cl2y.append(y[i])
    elif d3 < d1 and d3 < d2:
        print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 3")
        cl3x.append(x[i])
        cl3y.append(y[i])

centr1[0] = x[n[0]]
centr1[1] = y[n[0]]
centr2[0] = x[n[1]]
centr2[1] = y[n[1]]
centr3[0] = x[n[2]]
centr3[1] = y[n[2]]
        
a = 0
while a < 50:
    i  = 0
    Med1x = 0
    Med1y = 0
    Sum1x = 0
    Sum1y = 0
    while i < len(cl1x):
        Sum1x += cl1x[i]
        Sum1y += cl1y[i]
        Med1x = Sum1x/len(cl1x) #media coordonatelor x ale punctelor din clusterul 1
        Med1y = Sum1y/len(cl1x) #media coordonatelor y ale punctelor din clusterul 1
        i = i + 1
    Med1.append(Med1x) # x-ul noului centroid este adaugat in lista
    Med1.append(Med1y) # y-ul noului centroid este adaugat in lista
    print("Media cluster-ului 1 este ", Med1)

    i  = 0
    Med2x = 0
    Med2y = 0
    Sum2x = 0
    Sum2y = 0
    while i < len(cl2x):
        Sum2x += cl2x[i]
        Sum2y += cl2y[i]
        Med2x = Sum2x/len(cl2x) #media coordonatelor x ale punctelor din clusterul 2
        Med2y = Sum2y/len(cl2x) #media coordonatelor y ale punctelor din clusterul 2
        i = i + 1
    Med2.append(Med2x) # x-ul noului centroid este adaugat in lista
    Med2.append(Med2y) # y-ul noului centroid este adaugat in lista
    print("Media cluster-ului 2 este ", Med2)

    i  = 0
    Med3x = 0
    Med3y = 0
    Sum3x = 0
    Sum3y = 0
    while i < len(cl3x):
        Sum3x += cl3x[i]
        Sum3y += cl3y[i]
        Med3x = Sum3x/len(cl3x) #media coordonatelor x ale punctelor din clusterul 3
        Med3y = Sum3y/len(cl3x) #media coordonatelor y ale punctelor din clusterul 3
        i = i + 1
    Med3.append(Med3x) # x-ul noului centroid este adaugat in lista
    Med3.append(Med3y) # y-ul noului centroid este adaugat in lista
    print("Media cluster-ului 3 este ", Med3)
    
    
    del cl1x[:] #stergem clusterele pentru a fi refacute mai jos
    del cl1y[:]
    del cl2x[:]
    del cl2y[:]
    del cl3x[:]
    del cl3y[:]
    
    for i in range (0,9):
        d1 = math.sqrt((x[i] - Med1[0])**2 +(y[i] - Med1[1])**2) #calcularea distantelor la noii centroizi
        d2 = math.sqrt((x[i] - Med2[0])**2 +(y[i] - Med2[1])**2)
        d3 = math.sqrt((x[i] - Med3[0])**2 +(y[i] - Med3[1])**2)
        if d1 < d2 and d1 < d3:
            print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 1")
            cl1x.append(x[i])
            cl1y.append(y[i])
        elif d2 < d1 and d2 < d3:
            print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 2")
            cl2x.append(x[i])
            cl2y.append(y[i])
        elif d3 < d1 and d3 < d2:
            print("Punctul ","[",x[i],",",y[i],"]","apartine cluster-ului 3")
            cl3x.append(x[i])
            cl3y.append(y[i])
    if centr1  == Med1 and centr2 == Med2 and centr3 == Med3:
        break;
        
    
    centr1 = [0,0]
    centr2 = [0,0]
    centr3 = [0,0]
    
    centr1[0]= Med1[0]
    centr1[1]= Med1[1]
    centr2[0]= Med2[0]
    centr2[1]= Med2[1] 
    centr3[0]= Med3[0]
    centr3[1]= Med3[1] 
        
           
    
    del Med1[:]
    del Med2[:]
    del Med3[:]
    a = a + 1
```
